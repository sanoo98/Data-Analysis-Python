```python
#AI project
import numpy as np
from math import inf
from random import choice
import os
import time

SIZE = 3
h = -1
ai = 1
board = np.zeros(shape=(SIZE, SIZE), dtype=int)
# printBoard(board)

#display the board on terminal
def printBoard(board):
  print(" ", end='')
  for j in range(0, 4*np.size(board, 1)):

    print("-", sep='-', end='')
  print("-")
  for i in range(0, np.size(board, 0)):
    for j in range(0, np.size(board, 1)):
      print(" |", board[i][j], sep=' ', end='')
    print(" |\n ", end='')
    for j in range(0, 4*np.size(board, 1)):
      print("-", sep='-', end='')
    print("-")

def eval(board):
  if wins(board, ai):
    score = +1
  elif wins(board, h):
    score = -1
  else:
    score = 0
  return score

def wins(board, player):
  rowsums = np.sum(board, 0)
  colsums = np.sum(board, 1)
  pdiagsum = np.sum(np.diag(board))
  for i in range(0, np.size(rowsums)):
    if rowsums[i] >= SIZE or colsums[i] >= SIZE:
      return True
  if pdiagsum >= SIZE:
    return True


def gameOver(board):
  return wins(board, h) or wins(board, ai)


def emptyCells(board):
  cells = np.where(board == 0)
  return cells


def validMove(x, y):
  cells = emptyCells(board)
  if x in cells[0][:] and y in cells[1][:] :
    return True
  else:
    return False


def setMove(x, y, player):
  if validMove(x, y):
    board[x][y] = player
    return True
  else:
    return False


def minimax(board, depth, player):
  if player == ai:
    best = [-1, -1, -inf]
  else:
    best = [-1, -1, +inf]

  if depth == 0 or gameOver(board):
    score = eval(board)
    return [-1, -1, score]

  cells = emptyCells(board)
  for i in range(0, np.size(cells, 0)):
    x, y = cells[0][i], cells[1][i]
    board[x][y] = player
    score = minimax(board, depth - 1, -player)
    board[x][y] = 0
    score[0], score[1] = x, y

    if player == ai:
      if score[2] > best[2]:
        best = score  # max value
    else:
      if score[2] < best[2]:
        best = score  # min value
  return best


def ai_turn(ai_c, h_c):
  depth = np.size(emptyCells(board), 1)
  if depth == 0 or gameOver(board):
    return

  os.system("clear")
  print(f'Computer turn [{ai_c}]')
  printBoard(board)

  if depth == 9:
    x = choice([0, 1, 2])
    y = choice([0, 1, 2])
  else:
    move = minimax(board, depth, ai)
    x, y = move[0], move[1]

  setMove(x, y, ai)
  time.sleep(1)
    
    
def human_turn(ai_c, h_c):
  depth = len(emptyCells(board))
  if depth == 0 or gameOver(board):
    return

  os.system("clear")
  print(f'Human turn [{h_c}]')
  printBoard(board)

  if depth == 9:
    x = choice([0, 1, 2])
    y = choice([0, 1, 2])
  else:
    move = minimax(board, depth, h)
    x, y = move[0], move[1]

  setMove(x, y, h)
  time.sleep(1)


h_choice = 'X'
c_choice = 'O'
first = 'N'

# Main loop of this game
while len(emptyCells(board)) > 0 and not gameOver(board):
    if first == 'N':
        ai_turn(c_choice, h_choice)
        first = ''

    human_turn(c_choice, h_choice)
    ai_turn(c_choice, h_choice)

if wins(board, h):
    os.system('clear')
    print(f'Human turn [{h_choice}]')
    printBoard(board)
    print('YOU WIN!')
elif wins(board, ai):
    os.system('clear')
    print(f'Computer turn [{c_choice}]')
    printBoard(board)
    print('YOU LOSE!')
else:
    os.system('clear')
    printBoard(board)
    print('DRAW!')
```
